```python
from tensorflow.keras.models import load_model 
```

```python
def load_cnn(model_name, path):
    '''
    Function used to load a CNN model saved in h5 format
    Input: model directory, path where there is the folder Results with the model
    Output: model loaded
    '''

    def f1(y_true, y_pred):
        '''
        Function used to calculate the F1 score metric
        '''

        def recall(y_true, y_pred):
            '''
            Function used to calculate the Recall metric
            '''
            true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
            possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
            recall = true_positives / (possible_positives + K.epsilon())
            return recall

        def precision(y_true, y_pred):
            '''
            Function used to calculate the Precision metric
            '''
            true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
            predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
            precision = true_positives / (predicted_positives + K.epsilon())
            return precision

        precision = precision(y_true, y_pred)
        recall = recall(y_true, y_pred)
        
        return 2*((precision*recall)/(precision+recall+K.epsilon()))

    # Load model
    model = load_model(path + model_name + ".h5", 
                       custom_objects = {'f1': f1})

    return model
```

```python
model_name = 'cnn_model'
model = load_cnn('Results/cnn_model', "../")
```

```python
# serialize model to JSON
model_json = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("model.h5")
print("Saved model to disk")
```
